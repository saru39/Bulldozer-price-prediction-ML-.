```python
df['auctioneerID']=df['auctioneerID'][(df['auctioneerID']>0)& (df['auctioneerID']<6)]
```

```python
df['auctioneerID']=df['auctioneerID'].fillna(value=df.auctioneerID.mean())
```

```python
df['YearMade']=df['YearMade'][(df['YearMade']>1970) & (df['YearMade']<2015)]
```

```python
from datetime import datetime

def extract_date_time(string):
    datetime_obj=datetime.strptime(string, "%Y-%m-%d %H:%M:%S")
    date=datetime_obj.date()
    month=datetime_obj.month
    year=datetime_obj.year
    time=datetime_obj.time()
    return date,month,year,time

column_name='saledate'
dates=[]
months=[]
years=[]
times=[]

for value in df[column_name]:
    date,month,year,time=extract_date_time(str(value))
    dates.append(date)
    months.append(month)
    years.append(year)
    times.append(time)

df['Date'] = dates
df['Month'] = months
df['Year'] = years
df['Time'] = times

print(df)
```
